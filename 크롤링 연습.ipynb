```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'http://movie.daum.net/moviedb/grade?movieId=94074&type=netizen&page=1'
webpage = urlopen(url)
```

```python
source = BeautifulSoup(webpage, 'html5lib')
review = source.find('p', {'class':'desc_review'})
print(review.get_text().strip())
```

```python
#모든 댓글 크롤링_ source.find_all()로 바꾸깅
base_url = 'http://movie.daum.net/moviedb/grade?movieId=2725&type=netizen&page={}'
 
for n in range (20) :
    url = base_url.format(n+1)
    webpage = urlopen(url)
    source = BeautifulSoup(webpage, 'html5lib')
    reviews = source.find_all('p', {'class':'desc_review'})
    
    for review in reviews :
        print(review.get_text().strip())
```

```python
#위의 모든 댓글을 txt 파일로 자동 저장
review_list = []
for review in reviews :
    review_list.append(review.get_text().strip())
    
file = open('data1.txt', 'w', encoding = 'utf-8')
 
for review in review_list:
    file.write(review + '\n')
    
file.close() #저장 경로는 이 주피터가 열린 폴더 안에 자동저장됨.
```
