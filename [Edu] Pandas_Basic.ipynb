```python
#그룹바이(Groupby): 정말 많이 써요!
titanic_group=df.groupby('Pclass')
titanic_group 
```

```python
df.groupby('Pclass')['Age'].agg([max,min])
```

```python
##컬럼별로 적용하는 agg함수를 다르게 해봅시다
agg={'Age': 'mean', 'SibSp': 'sum', 'Fare': 'max'}
df.groupby('Pclass').agg(agg)
```

```python
#결측치 처리: 전처리의 기본!
##결측치(Missing data)란? 값이 없는, 즉 NULL 값을 의미합니다. NaN으로 표시
##결측치가 있는지 항상 전처리의 처음에 확인해야 하고, 이를 다른 값으로 대체해야 합니다.

#결측치 확인 : isna()
df.isna().head()
```

```python
df.isna().sum()
```

```python
#결측치 대체 : fillna()
df['Cabin']=df['Cabin'].fillna('C000')
df.head()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df.isna().sum()
```

```python
#결측치 삭제 : dropna()
##how 옵션엔 'all'(모든 컬럼에 NaN값이 있는 행 drop), 'any'(하나라도 NaN값이 있는 행 drop)
df=df.dropna(how='any')
df.isna().sum()
```

```python
#DataFrame 병합하기 : pd.concat() & pd.merge()
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']},
                  index = [0,1,2,3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                   'B': ['B4', 'B5', 'B6', 'B7'],
                   'C': ['C4', 'C5', 'C6', 'C7'],
                   'D': ['D4', 'D5', 'D6', 'D7']},
                  index = [4,5,6,7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                   'B': ['B8', 'B9', 'B10', 'B11'],
                   'C': ['C8', 'C9', 'C10', 'C11'],
                   'D': ['D8', 'D9', 'D10', 'D11']},
                  index = [8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
#pd.concat() 디폴트: axis=0(행 방향 결합), 합집합
result=pd.concat([df1,df2,df3])
result
```

```python
result2 = pd.concat([df1, df2, df3], keys = ['x', 'y','z']) #keys라는 인덱스값을 주면서 df들 병합
result2
```

```python
result2.index
```

```python
##axis=1(열 방향 결합)
df4 = pd.DataFrame({'B':['B2','B3','B6','B7'],
                  'D':['D2','D3','D6','D7'],
                  'F':['F2','F3','F6','F7']},
                   index = [2,3,6,7])
result3=pd.concat([df1,df4], axis=1)
result3
```

```python
##join='inner'(교집합)
result4=pd.concat([df1,df4], axis=1, join= 'inner')
result4
```

```python
#pd.merge() 디폴트: 교집합
G = pd.DataFrame({'key': ['K0','K4','K2','K3'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})
H = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
G
```

```python
H
```

```python
pd.merge(G,H) #따로 옵션을 주지 않아도, 공통된 컬럼인 'key'를 가지고 교집합으로 반환
```

```python
pd.merge(G,H, how='outer') #합집합으로 반환하고 싶다면 how='outer'
```
